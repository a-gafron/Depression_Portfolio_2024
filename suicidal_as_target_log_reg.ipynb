```python
def data_processing (dataframe):
    dataframe.gender = dataframe.gender.map({'male':1, 'female':0})
    dataframe['depressiveness'] = dataframe['depressiveness'].astype(int)
    dataframe['suicidal'] = dataframe['suicidal'].astype(int)
    dataframe['depression_diagnosis'] = dataframe['depression_diagnosis'].astype(int)
    dataframe['depression_treatment'] = dataframe['depression_treatment'].astype(int)
    dataframe['anxiousness'] = dataframe['anxiousness'].astype(int)
    dataframe['anxiety_diagnosis'] = dataframe['anxiety_diagnosis'].astype(int)
    dataframe['anxiety_treatment'] = dataframe['anxiety_treatment'].astype(int)
    dataframe['sleepiness'] = dataframe['sleepiness'].astype(int)
    return dataframe

df_really_clean = data_processing(df_cleaned)
```
